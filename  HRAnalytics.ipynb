**What is the attrition rate  of employee leaving the company having project less 3**

```python
lessProjAttrition = df.loc[(df['numberOfProjects']<3) & (df['left'] == 1)]
print(f'The number of people: {len(lessProjAttrition)}')
print(f'Percentage: {(len(lessProjAttrition)/11991)*100:.2f}%')
```

**What is the attrition rate  of employee leaving the company having project greater and equals 5**

```python
# 1: 9.14
# 2: 7.16 ---> Correct
more_project = df.loc[(df['numberOfProjects']>=5) & (df['left'] == 1)]
print(f'The number of people: {len(more_project)}')
print(f'Percentage: {(len(more_project)/11991)*100:.2f}%')
```

**The number of project is not very influencial factor to determine if an employee is going to leave a company or not**
